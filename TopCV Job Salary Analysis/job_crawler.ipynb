```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime
```

```python
headers= {'User=Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
raw_data = []
blank = ''
i=0
file = open('/content/drive/MyDrive/OLAP + Mining/dataset/link/link_all_3.txt', 'r', encoding='utf-8')
while (i<10):
    i+=1
    line = file.readline()
    if not line:
        break
    url = line.strip()
    response = requests.get(url, headers)
    if response.status_code == 200:
      soup = BeautifulSoup(response.content, 'html.parser')
      category = soup.find('div', attrs={"class": "ctn-breadcrumb-detail"})
      title = soup.find('div', attrs= {"class": "box-info-job"})
      info = soup.find('div', attrs={'class': 'box-info'})
      address = soup.find('div', attrs={'class': 'box-address'})
      description = soup.find('div', attrs={'class': 'job-data'})
      #keyword = soup.find('div', attrs={'class': 'box-key-word'})
      company_info = soup.find('div',attrs={'class': 'box-info-company box-white'})


      temp = []
      try:
        soup_category = BeautifulSoup(str(category), 'html.parser')
        category_elements = soup_category.findAll('a', attrs={'class': 'text-highlight bold'})
        temp.append(category_elements[1].get_text())
      except:
        temp.append(blank)

      title_soup = BeautifulSoup(str(title), 'html.parser')
      try:
        title_job = title_soup.find('h1', attrs={'class': 'job-title text-highlight bold'})
        temp.append(title_job.get_text())
      except:
        temp.append(blank)
      try:
        title_company = title_soup.find('div', attrs={'class': 'company-title'})
        temp.append(title_company.get_text())
      except:
        temp.append(blank)
      try:
        title_deadline = title_soup.find('div', attrs={'class': 'job-deadline'})
        temp.append(title_deadline.get_text())
      except:
        temp.append(blank)

      info_soup = BeautifulSoup(str(info), 'html.parser')
      info_elements = info_soup.find_all('span')
      for element in info_elements:
        try:
          temp.append(element.get_text())
        except:
          temp.append(blank)

      address_soup = BeautifulSoup(str(address), 'html.parser')
      try:
        address_elements = address_soup.find('div', attrs={'style': 'margin-bottom: 10px'})
        temp.append(address_elements.get_text())
      except:
        temp.append(blank)

      des_soup = BeautifulSoup(str(description), 'html.parser')
      try:
        des_elements= des_soup.find_all('div',{'class': 'content-tab'})
        for element in des_elements:
          temp.append(element.get_text())
      except:
        temp.append(blank)

      company_info_soup = BeautifulSoup(str(company_info),'html.parser')
      try:
        co_info_elements = company_info_soup.find_all('span')
        temp.append(co_info_elements[1].get_text())
        temp.append(co_info_elements[2].get_text())
      except:
        temp.append(blank)

      company_job_soup = BeautifulSoup(str(company_info),'html.parser')
      try:
        co_job_elements = company_job_soup.find('div',attrs={'class':'box-job-company'})
        co_job_elements_2 = co_job_elements.find_all('div', attrs={'class': 'box-item'})
        temp_2 =[]
        for element in co_job_elements_2:
          temp_2.append(element.get_text())
      except:
        temp_2 = ''
      temp.append(temp_2)

      raw_data.append(temp)

file.close()
now = datetime.datetime.now()
print('Crawl Job completed:', now.strftime('%Y-%m-%d %H:%M:%S'))
```

```python
data_headers =['category', 'job_name','company_name','job_deadline',
               'salary','num_recruits','working_form','level',
               'gender','exp','job_address','job_description',
               'job_require','job_benefit','company_size','company_address',
               'company_other_job']
```
